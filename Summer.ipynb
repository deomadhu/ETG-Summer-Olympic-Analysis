# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
```

1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d=df.loc[(df['Medal']=='Gold')]
d.groupby('Sport').count()['Medal'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```

3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```

4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```

5. Which player has won most number Gold Medals of medals? (Top 5)

```python
d=df.loc[(df['Medal']=='Gold')]
d.groupby('Athlete').count()['Medal'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```

6. In which year India won first Gold Medal in Summer Olympics?

```python
d=df.loc[(df['Country']=='IND') & (df['Medal']=='Gold')]['Year']
first_year=d.iloc[0]
first_year
```

7. Which event is most popular in terms of number of players? (Top 5)

```python
df.groupby('Event').count()['Athlete'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```

8. Which sport is having most female Gold Medalists? (Top 5)

```python
d=df.loc[(df['Medal']=='Gold') & (df['Gender']=='Women')]
d.groupby('Sport').count()['Gender'].sort_values(ascending = False).plot.bar(figsize = (6,5)).set_xlim(0,4)
```
